```python
import os
import torch
import torch.nn as nn
import torchvision
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import datasets, transforms

import datetime
from PIL import Image
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

## READING DATA

```python
transform = transforms.Compose([
    transforms.Resize((128,128)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Put datset path here
path_data = './dataset/'

class_names = []

for __class_name in os.listdir(path_data):
    class_names.append(__class_name)

dataset = datasets.ImageFolder(path_data, transform)
train_size = int(0.8 * len(dataset))  # 80% for training
test_size = len(dataset) - train_size  # 20% for testing

train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=True)
```

```python
class ConvNeuralNet(nn.Module):
    def __init__(self, num_classes):
        super(ConvNeuralNet, self).__init__()
        self.relu = nn.ReLU()                                                               # input:    128, 128, 3
        self.conv1 = nn.Conv2d(3, 10, kernel_size=6, padding=2, stride=2)                   #           64, 64, 10
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)                                  #           32, 32, 10
        self.conv2 = nn.Conv2d(10, 18, kernel_size=4, padding=2, stride=2)                  #           17, 17, 18
        self.pool2 = nn.MaxPool2d(kernel_size=3, stride=2)                                  #           8, 8, 18
        self.fc1 = nn.Linear(in_features=(8*8*18), out_features=128)                        
        self.fc2 = nn.Linear(128, num_classes)
    def forward(self, inputs):
        # 1st Conv layer
        x = self.conv1(inputs)
        x = self.relu(x)
        x = self.pool1(x)
        # 2nd Conv layer
        x = self.conv2(x)
        x = self.relu(x)
        x = self.pool2(x)
        # Flatten
        x = x.view(x.size(0), -1)
        # Fully Connected layers
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

## TRAINING

```python
def getModelName():
    current_time = datetime.datetime.now()
    time_format = "%Y-%m-%d_%H-%M-%S"  # Adjust the format as needed
    formatted_time = current_time.strftime(time_format)
    model_name = f"models/model_{formatted_time}.pt"
    return model_name
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(torch.cuda.is_available())
print(device)

# Define hyperparameters
LEARNING_RATE = 0.003
NUM_EPOCHS = 3
NUM_CLASSES = 3

# LEARNING_RATE = float(input("Enter Learning rate for the model: "))
# NUM_EPOCHS = int(input("Enter number of Epoch: "))

# Model and Others
model = ConvNeuralNet(num_classes=NUM_CLASSES).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)

# Training
total_step = len(train_loader)
print(f'No. of total steps: {total_step}')
print("starting training...")

losses = []

for epoch in range(NUM_EPOCHS):
    total_loss = 0.0
    for (images, labels) in tqdm(train_loader, desc=f"Epoch [{epoch}/{NUM_EPOCHS}]", unit="iteration"):
        images = images.to(device)
        labels = labels.to(device)

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)
        total_loss += loss.item()

        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    loss_per_epoch = total_loss / total_step
    losses.append(loss_per_epoch)

# Save the model
print("stoping training...")
torch.save(model.state_dict(), getModelName())
print("training stopped.")
print(f'Model saved as {getModelName()}')
```

## TESTING MODEL

```python
# Test the model
model = ConvNeuralNet(NUM_CLASSES).to(device)
# model_name = input("Enter model to load: ")
models = os.listdir('models/')
model_name = models[-1]
model.load_state_dict(torch.load('models/' + model_name))
print(f'"{model_name}" loaded.')

model.eval()
with torch.no_grad():
    correct = 0
    total = 0
    predictions = []
    true_labels = []
    for images, labels in tqdm(test_loader):
        images = images.to(device)
        labels = labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        predictions.extend(predicted)
        true_labels.extend(labels)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    print(f'Test Accuracy: {(100 * correct / total):.2f}%')
```

```python
# PRINT METRICS
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

cf = confusion_matrix(predictions, true_labels)
cf = cf.T

# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cf, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.ylabel('Predicted Label')
plt.xlabel('True Label')
plt.show()


plt.plot(np.arange(1, len(losses)+1, 1), losses, '-o')
plt.title('Train Loss vs Epoch')
plt.ylabel('Train Loss')
plt.xlabel('Epoch')
plt.xticks(np.arange(1, len(losses)+1, 1))
plt.show()
```
